```python
# real method
def find_largest_number(number_list: list[float]) -> float:
    """
    Function to find the largest number in the list

    Args:
        number_list (list[float]): A list containing numeric values (int or float)

    Returns:
        float: The largest number in the list

    Raises:
        TypeError: If input is not a list or contains non-numeric values.
        ValueError: If the list is empty.
    """

    # validate that input is a list
    if not isinstance(number_list, list):
        raise TypeError("Input must be a list of numbers (int or float).")

    # validate that list is not empty
    if not number_list:
        raise ValueError("List cannot be empty.")

    # validate that list elements are numeric
    for num in number_list:
        if not isinstance(num, (int, float)):
            raise TypeError("All elements must be numeric (int or float).")

    largest = number_list[0]
    for num in number_list:
        if num > largest:
            largest = num

    return largest


# main execution
if __name__ == "__main__":
    try:
        # taking input from user
        user_input = input("Enter numbers separated by space (int or float only): ")

        # converting user input into list
        number_list = [float(item) for item in user_input.split()]

        # calling the function
        result = find_largest_number(number_list)

        # display the result
        print("Largest number in the list is:", result)

    except (TypeError, ValueError) as error:
        print("Error:", error)

    except Exception as e:
        print("Unexpected error:", e)
```
