```python
# question 4:- Find Prime Numbers Between 100 and 200 

def is_prime_number(number: int) -> bool:
    """
    Function to check whether a number is prime or not.

    Args:
        number (int): The number to check.

    Returns:
        bool: True if the number is prime, False otherwise.

    Raises:
        TypeError: If input is not an integer.
        ValueError: If number is less than 2.
    """

    # Validate input type
    if not isinstance(number, int):
        raise TypeError("Input must be an integer")

    # Prime numbers start from 2
    if number < 2:
        raise ValueError("Prime numbers are greater than or equal to 2")
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False

    return True

if __name__ == "__main__":
    try:
        prime_number = []  #  for making a list with the names of prime numbers
        lower_limit = input("Enter lower limit :")
        upper_limit = input("Enter upper limit : ")
# validate for handling error 
        if not lower_limit.strip().isdigit():
            raise TypeError("limit must be a valid integer (no decimals or letters)")
        if not upper_limit.strip().isdigit():
            raise TypeError("limit must be a valid integer (no decimals or letters)")   
        lower_limit = int(lower_limit)  # if any user provides something, change into int
        upper_limit = int(upper_limit)
        
        for i in range(lower_limit, upper_limit):   # if we need without user define we modiy it (100,200)
            if is_prime_number(i):
                print(f" {i} is prime number")
                prime_number.append(i)
        print(prime_number)
    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
     
    
```
