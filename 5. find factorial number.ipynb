```python
# real method 
def factorial_finder(num: int) -> int:
    """
    Calculate the factorial of a non-negative integer.

    Parameters:
        num (int): The number for which factorial is required.
    Returns : 
       int: Factorial of the given number.
    Raises:
         ValueError: If num is negative.
         TypeError: If num is not an integer.
    """
    if isinstance(num,bool):
        raise TypeError("Boolean values are not allowed")
    # check type of num
    if not isinstance(num,int):
        raise TypeError("Input must be an integer")
    # negative check
    if num < 0:
         raise ValueError("Factorial is not defined for negative numbers")
    
    factorial = 1
    for i in range(1,num+1):
        factorial = factorial * i   
    return factorial
# main execution
if __name__ == "__main__":
    try:
       number = input("Enter your number to find factorial : ")
       number = int(number)
       print("Factorial : " , factorial_finder(number))
    except ValueError as e:
        print("Input must be a positive integer number")
    except TypeError as e:
        print("Error :",e)
```
